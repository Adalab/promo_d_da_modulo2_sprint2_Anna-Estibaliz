Para esto trabajaremos con el fichero generado en Limpieza I.

- Columnas loan, housing y default: estas columnas contienen valores √∫nicos de 0 y 1. Esto puede parecer poco intuitivo a la hora de la extracci√≥n de conclusiones y en las visualizaciones. El objetivo de este ejercicio es que cambies los valores n√∫mericos por "Si" y "No". A que corresponde cada uno de los valores lo ten√©is en el pair de Limpieza I.


```python
df['default'] = pd.cut(df['default'], 2, labels= ['Si', 'No'])
```

```python
df['housing'] = pd.cut(df['housing'], 2, right=True, labels= ['Si', 'No'])
df['loan'] = pd.cut(df['loan'], 2, right=True, labels= ['Si', 'No'])
```


- Para la columna de education, si nos fijamos en sus valores √∫nicos veremos que tenemos puntos. El objetivo de este ejercicio es que quit√©is los puntos de esos valores y los reemplac√©is por espacios.


```python
df['education'] = df['education'].str.replace('.', ' ')
```


- Para la columna job, hay un valor √∫nico que esta abreviado (admin.), cambiad la abreviatura por el nombre completo.


- La columna month_day_week tiene una lista que contiene informaci√≥n muy diferente. El objetivo de este ejercicio es separar esta columna en dos nuevas, donde tengamos los meses en una columna y los d√≠as de la semana en otra.


```python
df[['month', 'weekday']] = df['month_day_week'].str.split(',', n=2, expand=True)
```

```python
df['month'] = df['month'].str.replace('[', ' ')
```

```python
df['weekday'] = df['weekday'].str.replace(']', ' ')
```
